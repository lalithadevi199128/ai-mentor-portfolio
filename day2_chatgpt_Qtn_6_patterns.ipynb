Step 5: How to Calculate Big-O Quickly

Rule 2: Drop Smaller Terms

Example:

O(n²)

Rule 3: Sequential Loops Add
for(i=0;i<n;i++)   // O(n)

O(n × n) = O(n²)

“What is recursion?”

Don’t jump into Fibonacci immediately. Start clean and simple:

“Recursion is a technique where a function calls itself to solve smaller instances of the same problem.”

That’s your core definition.

Step 1: Intuition (Very Important)

Think of recursion like breaking a big problem into smaller same-type problems.

Example:
You want to go down 5 stairs.

What do you do?

Go to step 4
Then step 3
Then step 2…

Same action repeated → smaller version each time.

That’s recursion thinking.

Step 2: Two Mandatory Parts

Every recursive function MUST have:

1. Base Case (Stopping condition)

Without this → infinite loop → stack overflow.

Example:

if n == 0:
    return 1
2. Recursive Case

Function calling itself.

return n * factorial(n-1)
Step 3: Classic Example (Factorial)
def factorial(n):
    if n == 0:          # base case
        return 1
    return n * factorial(n-1)   # recursive call
How it works (very important for interviews):

factorial(3)
→ 3 * factorial(2)
→ 3 * 2 * factorial(1)
→ 3 * 2 * 1 * factorial(0)
→ 3 * 2 * 1 * 1
→ 6

Step 4: What’s Happening Internally

Every function call is stored in memory (call stack).

So recursion uses extra memory.

That’s why:

Too many calls → stack overflow
Interviewers may ask about space complexity
Step 5: Time Complexity Example

Factorial:

Calls function n times → O(n)

Fibonacci (naive recursion):

def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

This is:

O(2^n)

Very slow → interview trap question.

Step 6: When to Use Recursion

Use recursion when problem has:

Repeated subproblems
Tree structure
Divide-and-conquer pattern

Common examples:

Factorial
Fibonacci
Binary Search
Tree traversals
Backtracking (subsets, permutations)
Step 7: Recursion vs Loop (Interview Favorite)
Recursion	Iteration
Uses function calls	Uses loops
Uses extra memory (stack)	Memory efficient
Elegant for trees/problems	Faster in many cases

Good answer in interview:

“Recursion is more intuitive for hierarchical problems, but iteration is more space-efficient.”

Step 8: Common Mistakes Students Make

Wrong recursive call
Not reducing problem size
Not analyzing complexity

“Recursion is when a function solves a problem by calling itself on smaller inputs. It requires a base case to stop and a recursive case to reduce the problem. For example, factorial uses recursion by multiplying n with factorial of n-1 until it reaches 0.”

What is a stack?

Good, this is exactly how interviewers build the conversation:

Let’s make it clear and placement-ready.

Simple Definition (Say this in interview)

“A stack is a linear data structure that follows the LIFO principle — Last In, First Out.”

That one line is must-know.

Step 1: Real-Life Analogy

Think of a stack of plates 🍽️

You place plates one over another
The last plate you put → is the first one you remove

That is exactly how a stack works.

Step 2: Basic Operations

Only two main operations (interviewers love this):

1. Push

Add element to top

Push(10), Push(20)
Stack → [10, 20]
2. Pop

Remove element from top

Pop()
Stack → [10]
Extra (good to mention)
Peek / Top → see top element
isEmpty() → check if stack is empty
Step 3: Example
Push(1)
Push(2)
Push(3)

Stack looks like:
Top → 3
        2
        1

Now:

Pop() → 3
Pop() → 2
Pop() → 1

Order becomes reversed.

Step 4: Where Stack is Used (Very Important)

Interviewers often ask this.

1. Function Calls (Call Stack)

When you use recursion:

def fun(n):
    if n == 0:
        return
    fun(n-1)

Each function call is stored in stack.

So stack manages:

which function is running
where to return
2. Expression Evaluation

Examples:

(a + b) * c
Infix → Postfix conversion
3. Undo/Redo Operations
Text editors
Browser history
4. Parenthesis Matching

Check if:

( [ { } ] )

is valid or not.

Step 5: Time Complexity

All stack operations are:

O(1)

Push → constant
Pop → constant

Good point to mention in interviews.

Step 6: Implementation

Two ways:

1. Using Array
Fixed size
Fast
2. Using Linked List
Dynamic size
No overflow (until memory full)

Implement stack using array
Implement stack using two queues
Check balanced parentheses
Reverse a string using stack
Step 8: Connection to Recursion (Important)

Earlier we discussed recursion.

Now connect:

“Recursion internally uses a stack (call stack) to store function calls.”

That’s why:

Deep recursion → stack overflow
Each call → pushed into stack
Return → popped from stack

What is Big-O notation?

Let’s tighten this into a perfect interview answer + clear understanding.

What is Big-O Notation? (Say this first)

“Big-O notation is used to describe how the time or space complexity of an algorithm grows as the input size increases.”

If you say this clearly, interviewer knows you understand the concept.

Step 1: Why Do We Need It?

Suppose you write two programs:

One works fast for 10 inputs
Another works fast even for 10 lakh inputs

Which one is better?

👉 Big-O helps us compare algorithms independent of machine, language, or constants.

Step 2: Focus on Growth, Not Exact Time

Big-O does NOT tell:

exact seconds
exact number of steps

It tells:

“How fast the work increases when input (n) increases.”

Step 3: Example (Very Important)
Linear Search
for i in range(n):
    if arr[i] == key:
        return i

Worst case:

Check all elements → n operations

👉 O(n)

Binary Search

Each step halves the data.

👉 O(log n)

Even for large input, it's fast.

Step 4: Common Big-O Values (Must Remember)
Big-O	Meaning	Example
O(1)	Constant time	Access array element
O(log n)	Logarithmic	Binary search
O(n)	Linear	Loop through array
O(n log n)	Efficient sorting	Merge sort
O(n²)	Nested loops	Bubble sort
O(2ⁿ)	Exponential	Recursive Fibonacci
Step 5: Key Rules (Scoring Points in Exams)
1. Ignore constants
O(2n) → O(n)
2. Ignore smaller terms
O(n² + n) → O(n²)
3. Nested loops multiply
O(n × n) = O(n²)
4. Sequential loops add
O(n + n) = O(n)
Step 6: Best, Average, Worst Case

Interviewers may go here.

Example: Linear search

Best case → O(1) (element found first)
Worst case → O(n)

Usually, we talk about worst-case Big-O.

Step 7: One Strong Interview Answer

If asked directly:

“Explain Big-O notation”

Say:

“Big-O notation represents the upper bound of an algorithm’s time or space complexity, describing how it scales with input size. It helps compare algorithms by focusing on growth rate rather than exact execution time.”

That “upper bound” word gives extra weight.

Step 8: Common Mistake

Students say:

“Big-O is time taken by program”

❌ Not correct

Correct:

“It is the growth rate of time with respect to input size”
